# Movie Recommendation using Item-Based Collaborative Filtering

Data regading movies and ratings to it is taken from MovieLens dataset. Loading the data from u.data & u.item and merging them inorder to work with movie names instead of movie id.

```python
import pandas as pd

r_cols = ['user_id', 'movie_id','rating']
ratings = pd.read_csv('./ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('./ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
#Creating a matrix from it using pivot_table function
userRating = ratings.pivot_table(index=['user_id'], columns=['title'],values='rating')
userRating.head()
```

corr() method that will compute a correlation score for every column pair in the matrix! This gives us a correlation score between every pair of movies (where at least one user rated both movies - otherwise NaN's will show up.)

```python
corrMatrix = userRating.corr()
corrMatrix.head()
```

```python
corrMatrix = userRating.corr(method='pearson', min_periods=100)
#min_periods mean that atleast 100 people should have rated
corrMatrix.head()
```

let's produce some movie recommendations for user ID 0

```python
myRating = userRating.loc[0].dropna()
myRating.head()
```

Now let's go through the movies rated by user ID 0 one by one and recommend a list of movies based on the movies similar to the ones that were rated.

retrieve the list of similar movies from the correlation matrix. I'll then scale those correlation scores by how well the user rated the movie they are similar to, so movies similar to ones  liked count more than movies similar to ones hated

```python
similar_movie = pd.Series()
for i in range(0, len(myRating.index)):
    print("Adding similarities for "+ myRating.index[i]+"...")
    #Retriving similar movies to this rated one
    similar = corrMatrix[myRating.index[i]].dropna()
    similar = similar.map(lambda x: x*myRating[i])
    
    similar_movie = similar_movie.append(similar)

print("\n")    
print("sorting...")
similar_movie.sort_values(inplace = True, ascending = False)
similar_movie.head(10)
```

```python
#grouping the movies, so that they won't be repeated
similar_movie = similar_movie.groupby(similar_movie.index).sum()
```

```python
similar_movie.sort_values(inplace = True, ascending = False)
similar_movie.head(10)
```

```python
#removing the movies that the user has already rated
filteredSimilarity = similar_movie.drop(myRating.index)
filteredSimilarity.head(10)
```
